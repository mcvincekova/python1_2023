```python
import pandas

names = pandas.read_csv("https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/agregace-a-spojovani/excs/excs>studenti/jmena.csv")
students1 = pandas.read_csv("https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/agregace-a-spojovani/excs/excs>studenti/studenti1.csv")
students2 = pandas.read_csv("https://kodim.cz/cms/assets/kurzy/python-data-1/python-pro-data-1/agregace-a-spojovani/excs/excs>studenti/studenti2.csv")
```

```python
students2
```

```python
# Načti dva datové sety studentů do oddělených pandas DataFrame a pomocí funkce concat je spoj do jednoho setu.

students = pandas.concat([students1, students2], ignore_index=True)
students
```

```python
pandas.concat([students1, students2])
```

```python
# Pokud studentovi chybí ročník, znamená to, že již nestuduje. 
# Pokud mu chybí číslo skupiny, znamená to, že jde o dálkového studenta.
# Kolik studentů v datovém setu již nestuduje a kolik jsou dálkoví studenti?

remote_students = students[students["kruh"].isnull()]
# len(remote_students)
remote_students.info()
```

```python
no_longer_students = students[students["ročník"].isnull()]
no_longer_students
```

```python
# Vyčisti data od studentů, kteří nestudují nebo studují jen dálkově. Nadále budeme pracovat pouze s prezenčními studenty.
students = students.dropna(subset=["ročník", "kruh"])
students.info()
```

## Seminar 10 (d. - g.)

```python
# D) Zjisti, kolik prezenčních studentů je v každém z oborů.

students.groupby("obor")["jméno"].count()
```

```python
students["obor"].value_counts()
```

```python
# E) Zjisti průměrný prospěch studentů v každém oboru.

students.groupby("obor")["prospěch"].mean().sort_values(ascending=False)
```

```python
# F) Načti datový set s křestními jmény. Proveď join s tabulkou studentů tak, abychom věděli pohlaví jednotlivých studentů.

spojeny = pandas.merge(students, names[["jméno", "pohlaví"]], on = ["jméno"])
spojeny
```

```python
# G) Zjisti, zda na naší fakultě studují IT spíše ženy nebo spíše muži.

spojeny["pohlaví"].value_counts()
```

```python
spojeny.groupby("pohlaví").size()
```

```python
# Zmen data v stlpci 'pohlaví' tak, ze namiesto malych, bude obsahovat velke pismena - teda hodnoty 'Ž' a 'M' namiesto 'ž' a 'm'

spojeny["pohlaví"] = spojeny["pohlaví"].str.upper()
spojeny
```

```python
# Zisti pocet muzov a zien v jednotlivych rocnikoch

spojeny.groupby(["ročník", "pohlaví"])["příjmení"].count()
```

```python
# Vytvor slpec 'cele_meno', v ktorom bude ulozene cele meno studenta (jméno a příjmení); napr. 'Ludmila Šindlerová'

spojeny["cele_meno"] = spojeny["jméno"] + " " + spojeny["příjmení"]
spojeny
```
